```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Dataset raw url
url = 'https://raw.githubusercontent.com/viable-data-analyst/Sea_Level_Predictor/refs/heads/main/epa-sea-level.csv?token=GHSAT0AAAAAAC4GLH5VAVOABY5C4NRRH5SQZ3RLQMQ'
```

```python
# Read data from file
df = pd.read_csv(url, sep = ",")

# Scatter
plt.figure(figsize= (12,6))
plt.scatter(x= df['Year'], y= df['CSIRO Adjusted Sea Level'],
                color = 'blue', alpha= 0.5)

# Create first line of best fit
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(range(1880, 2051, 1), slope * range(1880, 2051, 1) + intercept, color= 'red')

# Create second line of best fit
slope, intercept, r_value, p_value, std_err = linregress(df.loc[df['Year']>= 2000][['Year','CSIRO Adjusted Sea Level']])
plt.plot(range(2000, 2051, 1), slope * range(2000, 2051, 1) + intercept, color= 'green')

# Add labels and title
plt.title('Rise in Sea Level')
plt.ylabel('Sea Level (inches)')
plt.xlabel('Year')
plt.show()
```
